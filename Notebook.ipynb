```python
import sqlite3 as sql

# Connect to the database. If it doesn't exist, it will be created.
connect = sql.connect("my_db.db")
cursor_object = connect.cursor()

# SQL to create the table
table_query = """
CREATE TABLE MYTABLE(
    ID INT NOT NULL,
    FNAME VARCHAR(30),
    LNAME VARCHAR(30),
    CITY VARCHAR(20),
    CCODE CHAR(10)
);"""

# Drop the table if it already exists to ensure a clean start
cursor_object.execute("DROP TABLE IF EXISTS MYTABLE")
cursor_object.execute(table_query)

# SQL to insert multiple values
add_values = """
INSERT INTO MYTABLE(ID, FNAME, LNAME, CITY, CCODE)
VALUES 
    (1, 'Haseeb', 'Sagheer', 'Barnala', '11'), 
    (2, 'Adeel', 'Sagheer', 'Bhimber', '22'),
    (3, 'Kabir', 'Sagheer', 'Dammam', '33');
"""
cursor_object.execute(add_values)

# --- THE MISSING CRITICAL STEP ---
# Commit the changes to permanently save them to the database file.
connect.commit()

# Close the connection to the database.


print("Database created and populated successfully!")
```

```python
view_data = """SELECT * FROM MYTABLE"""
cursor_object.execute(view_data)

# <Strong>Bonus: now write and execute an update statement that changes the Rav's CITY to MOOSETOWN</strong>

quote_data = '''UPDATE MYTABLE SET CITY = "MOOSTOWN" where FNAME = 'Haseeb';'''
cursor_object.execute(quote_data)
```
